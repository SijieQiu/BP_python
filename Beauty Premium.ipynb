Q1:What impact does wearing makeup have on the number of transfers?

*Identify the count of instances where the transfer value surpasses the average. Determine the distribution between makeup and non-makeup categories.

```python
import pandas as pd

exceeds_transfer_avg = BP_df[BP_df['transfer'] > BP_df['transfer'].mean()]

makeup_count = (exceeds_transfer_avg['treatment'] == 'makeup').sum()
nomakeup_count = (exceeds_transfer_avg['treatment'] == 'nomakeup').sum()

print(f"Count of 'makeup' instances exceeding transfer average: {makeup_count}")
print(f"Count of 'nomakeup' instances exceeding transfer average: {nomakeup_count}")
```

*To make it more clear

```python
import matplotlib.pyplot as plt
plt.bar(['makeup', 'nomakeup'], [makeup_count, nomakeup_count], color=['blue', 'orange'])
plt.xlabel('Treatment')
plt.ylabel('Count')
plt.title('Counts of Treatment Exceeding Transfer Average')
plt.show()
```

Q2:How does an increase in perceived attractiveness (IncreaseAttract) influence the perceived 'Trustworthiness' of individuals in a given context?

```python
import matplotlib.pyplot as plt

plt.scatter(BP_df['Trustworthiness'], BP_df['IncreaseAttract'], c='red', alpha=0.5)  
plt.xlabel('Trustworthiness')
plt.ylabel('Increase in Attractiveness')
plt.title('Influence of Trustworthiness on Increase in Attractiveness')
plt.show()
```

* This scatter plot above where the x-axis represents Trustworthiness and the y-axis represents the increase in attractiveness ("IncreaseAttract").This visualization aims to display any potential relationship or pattern between these two variables.



Q3:Does the use of makeup (makeupdummy) significantly impact the perceived attractiveness increase (IncreaseAttract) in the trust game experiment? ( based on the 10 randomly selected samples above)

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'player2': [16, 29, 30, 10, 15, 4, 12, 35, 5, 34],
    'makeupdummy': [0, 1, 0, 1, 0, 0, 0, 1, 0, 0],
    'IncreaseAttract': [1.50, -0.75, 3.75, 0.00, 1.50, 2.25, -0.75, 1.75, 2.75, 1.00]
}

import pandas as pd
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.boxplot(x='makeupdummy', y='IncreaseAttract', data=df)
plt.xlabel('Makeup (0: No Makeup, 1: Makeup)')
plt.ylabel('Increase in Attractiveness')
plt.title('Effect of Makeup on Increase in Attractiveness')
plt.show()
```

*The boxplot above compares the distribution of "IncreaseAttract" values between the makeup and non-makeup groups. The x-axis represents the makeup status (0: No Makeup, 1: Makeup), and the y-axis shows the increase in attractiveness.
